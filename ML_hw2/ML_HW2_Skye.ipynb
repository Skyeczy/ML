```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from random import shuffle
```

## Problem 2-(b) MLE

```python
#First create a random list with 6 1s and 4 0s.
n = 10
X = [0] * 4 + [1] * 6
shuffle(X)
# create list theta
theta = [0.01*i for i in range(100)]
```

```python
#calculate likelihood function for each theta
likelihood_list = []
for i in range(len(theta)):
    likelihood = 1
    for j in range(n):
        likelihood = likelihood * theta[i]**(int(X[j]))*(1-theta[i])**(1-int(X[j]))
#     likelihood_list= likelihood_list +[likelihood]                                                                
    likelihood_list.append(likelihood)
```

```python
#plot
plt.plot(theta, likelihood_list)
plt.xlabel("$\hat{ \\theta} $")
plt.ylabel("Value of likelihood function")
```

```python
likelihood_array = np.asarray(likelihood_list)
theta_hat_index = (-likelihood_array).argsort()[:1][0]
theta_hat = theta[theta_hat_index]
print('The value of \\theta_hat that maximizes the likelihood is', theta_hat)
```

## Problem 2d

```python
#function to calculate p(\theta)
def p_theta(theta):
    return (theta**2)*((1-theta)**2)/0.033333
#calculate likelihood function for each theta
likelihood_list = []
alpha_1 = sum(X)
alpha_0 = n-sum(X)
for i in range(len(theta)):
    likelihood = theta[i]**alpha_1
    likelihood = likelihood * (1-theta[i])**alpha_0
    likelihood = likelihood *  p_theta(theta[i])
    likelihood_list.append(likelihood)
#plot
plt.plot(theta, likelihood_list)
plt.xlabel("$\hat{ \\theta} $")
plt.ylabel("Value of MAP likelihood function")
```
